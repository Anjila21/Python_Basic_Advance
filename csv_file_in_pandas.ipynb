```python
# To show desired no of rows
data = pd.read_csv('Data.csv',nrows = 4)
data
```

```python
# To show desired number of columns

data = pd.read_csv('Data.csv', usecols = [1,2,3])
data
```

```python
# To Skip the rows

data = pd.read_csv('data.csv',skiprows=12) # skips 12 rows from start
data
```

```python
# To change the index of the file

data = pd.read_csv("data.csv",index_col ="Gender") #The index is now the gender
data
```

```python
# To change the header

data = pd.read_csv('Data.csv' ,header = 1) # index[1] becomes the header
data
```

```python
# To remove header

data = pd.read_csv("Data.csv", header=None) # Header is changed to default
data
```

```python
#slicing the column
data.loc[:,[0,1,2]]
```
