Access a groupofrows and columns by labels or boolean array.
Syntax: DataFrame.loc[]

```python
import pandas as pd
df = pd.read_csv("Data1.csv")
df
```

```python
df.loc[0]  #gives the value of '0' index
```

```python
df.loc[[3,5,7,9,11]]  #gives value of 3,5,7,9,11 index
```

```python
df.loc[9,'Experience']  #gives the values of corresponding column with respect to its index
```

```python
df.loc[df['Income']<45]  #prints the row of "Income" column whose value is less than 45
```

### iloc: Integer location-based indexing

```python
df.iloc[[0]]  #display the value of '0' index
```

```python
df.iloc[0:4,1:6] #displays the value of row from '0-3' index and column from'1-5' index
```
