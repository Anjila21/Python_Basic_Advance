Matplotlib is 2D and 3D plotting python library.With the help of Matplotlib we can create:
1. Histogram
2. Bar Chats
3. Scatterplost
4. Area Plot
5. Pie Plot
6. Error Charts
7. Power Spectra

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
days = np.arange(10,20)
temp = [30.4,40.1,23,15,20,23,21.33,22.3,20,23]
plt.plot(days,temp)
plt.axis([0,21,0,41])  #To define starting and ending point of the graph
plt.title("Kathmandu Temperature")  #To give title 
plt.xlabel("days")   # To give name to x-axis
plt.ylabel("temp")  #To give name to y-axis
plt.show()
```
