### Line Plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
days = np.arange(10,20)
temp = [30.4,40.1,23,15,20,23,21.33,22.3,20,23]
plt.plot(days,temp,color="m",marker='o',linestyle='--',linewidth=3,markersize=9)
plt.axis([0,21,0,41])  #To define starting and ending point of the graph
plt.title("Kathmandu Temperature")  #To give title 
plt.xlabel("days")   # To give name to x-axis
plt.ylabel("temp")  #To give name to y-axisk
plt.show()
```

### Alternative for above Code

```python
days = np.arange(10,20)
temp = [30.4,40.1,23,15,20,23,21.33,22.3,20,23]
plt.plot(days,temp,"mo--",linewidth=3,markersize=9) #Here we can directly give same parameter at once
plt.axis([0,21,0,41])  #To define starting and ending point of the graph
plt.title("Kathmandu Temperature",fontsize=14)  #To give title 
plt.xlabel("days",fontsize=12)   # To give name to x-axis
plt.ylabel("temp",fontsize=12)  #To give name to y-axisk
plt.show()
```

```python
days = np.arange(10,20)
temp = [30.4,40.1,23,15,20,23,21.33,22.3,20,23]
plt.plot(days,temp,"mo--",linewidth=3,markersize=9) #Here we can directly give same parameter at once
plt.axis([0,21,0,41])  #To define starting and ending point of the graph
plt.title("Kathmandu Temperature",fontsize=14)  #To give title 
plt.xlabel("days",fontsize=12)   # To give name to x-axis
plt.ylabel("temp",fontsize=12)  #To give name to y-axisk
plt.legend(['Temp Line'],loc=4)   #To give name to line
plt.grid(color='r',linestyle='-',linewidth=2)
plt.show()
```

### To Plot multiple graph

```python
days = np.arange(10,20)
temp_KTM = [30.4,40.1,23,15,20,23,21.33,22.3,20,23]
temp_BKT = [20.11,20.1,33,35,20,23,19.33,22.3,20,23]
plt.plot(days,temp_KTM,"mo--",linewidth=3,markersize=9,label="KTM Temp") #Here we can directly give same parameter at once
plt.plot(days,temp_BTK,"yo--",linewidth=3,markersize=9,label="BKT Temp")
plt.axis([0,21,0,41])  #To define starting and ending point of the graph
plt.title("Kathmandu and Bhaktapur Temperature",fontsize=14)  #To give title 
plt.xlabel("days",fontsize=12)   # To give name to x-axis
plt.ylabel("temp",fontsize=12)  #To give name to y-axisk
plt.legend(loc=4)
plt.grid(color='r',linestyle=':',linewidth=1)
plt.show()
```
