Pandas replace:
Values to the dataframe are replaced with othe values dynamically.
Syntex: DataFrame.replace()
Parameters: to_replace,value,inplace,limit,regex,method

```python
import pandas as pd
data=pd.read_csv("Data.csv")
data
```

```python
data.replace(to_replace='Rent',value="owned") #replaces the value with your own value
```

```python
data.replace({'Gender':"M"},"Other")  #replaces the column gender having 'M' value with "other"
```

```python
data.replace('[A-Za-z]',0,regex=True) #replaces all the string values with "0"
```

```python
data = pd.read_csv("Data.csv")
data.replace('M',method='bfill')
```
