```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
girl_age = np.random.randint(19,25,(50))
boy_age = np.random.randint ((22,30),(50))
print(girl_age)
print(boy_age)
```

```python
plt.hist(girl_age)
plt.title("Girls Student Age")
plt.xlabel("Students age Category")
plt.ylabel("No. Students Age")
plt.show()
```
