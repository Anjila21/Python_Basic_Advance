```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
```

```python
iris = pd.read_csv("C:\\Users\\Dell\\Desktop\\Anjila\\anjila_internship\\Python\\data_analysis\\iris\\IRIS.csv")
iris.head()
```

```python
iris['species'].value_counts()
```

```python
sns.FacetGrid(iris,hue='species',height=8).map(plt.scatter,'petal_length','petal_width').add_legend()
```

```python
flower = {'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
iris['species']=iris['species'].map(flower)
iris.head()
```

```python
X=iris[['sepal_length','sepal_width','petal_length','petal_width']].values
y=iris[['species']].values
y=np.ravel(y)
```

```python
model = SVC(kernel='linear',C=0.1)
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
expected= y
predicted = model.predict(X)
predicted
```

```python
from sklearn import metrics
print(metrics.classification_report(expected,predicted))
print(metrics.confusion_matrix(expected,predicted))
```
