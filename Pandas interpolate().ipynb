Pandas interpolate() function is basically used to fill NaN values in dataframe or series.

Syntax: DataFrame.interpolate() 
        Series.interpolate()
        
Parameters: mmethod,axis,limit,inplace,limit_direction,limit_area

```python
import pandas as pd
data = pd.read_csv("Data1.csv")
data
```

```python
data.interpolate(method="linear")  #interpolate function is only used in integer value as it is easiler to predict in numerical value
```

```python
type(data.Date[0])  #datetype of "Date" column is string so we have to convert it into data type
```

```python
data = pd.read_csv("Data1.csv",parse_dates=['Date'])  #converting the "Date" datatype from string to time
data
```

```python
type(data.Date[0]) #checking datatype
```

```python
data = pd.read_csv('Data1.csv',parse_dates=['Date'],index_col='Date')  #changes the index from 'id' to 'Date'
data
```

```python
data.interpolate(method='time')  #Nan values gets changes depending upon date attribute
```

```python
data = pd.read_csv("Data1.csv",index_col='Income')  #changes the index as "Income"
data
```

```python
data.interpolate(method='index')
```

```python
data.interpolate(method ="nearest") #fills the value corresponding to its nearest value
```

```python
data.dtypes
```

```python
data.interpolate(limit=3, limit_direction="backward") #fills the 3 empty shell from the backward with the corresponding backward value
```

```python
data.interpolate(limit=1,limit_direction='both') #uses the corresponding values from forward and backward in one shell
```

```python
data.interpolate(limit_area='inside') #fills NaN value inside
```
