Pandas gropby function is used to split the data into groups based on some criteria.
Three type of operations performed by groupby() are:
1. Splitting the object
2. Applying a function
3. Combining the result

```python
import pandas as pd
data= pd.read_csv('Data.csv')
data
```

```python
group = data.groupby(by='Education')
group
```

```python
group.groups  #groups the repeated value
```

```python
group1 = data.groupby('Mortgage').get_group(0)  #displays the rows who Mortgage value is '0'
group1
```

```python
group1.sum()  #add the all the values and gives total
```

```python
group1.agg(['sum','mean','median']) #To find sum,mean and median
```
