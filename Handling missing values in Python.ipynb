NaN --> Not a number

```python
import pandas as pd
data = pd.read_csv("Data.csv")
data
```

```python
# To find out if there exists any null value

data.isnull() #if true there exist null value
                #if false there doesnt exist null value
```

```python
# To show total number of null values in each column in numeric form

data.isnull().sum()
```

```python
# To find total number of null values in data 

data.isnull().sum().sum()  
```

```python
# To drop the corresponding rows having null values

data.dropna()
```

```python
data = pd.read_csv("Data.csv")
data.dropna(axis =1) # drops th column having null values
```

### how  = any ( it is default , it removes the rows and columns having null values)  how = all ( deletes a row or column if it consists all the values of the particular rows or columns as null


```python
data.dropna(axis=1 , how ='any')
```

```python
data.dropna(axis= 0 , how = 'all')  #no rows are dropped as there doesnt consist of row having all the null values
```

```python
data = pd.read_csv('Data.csv')
data.dropna(inplace = True)   #drops rows having null values
data
```
