```python
#missing data in percentage

null_var = data.isnull().sum()/data.shape[0]*100
null_var
```

```python
# removes all the column having missing data more than 20%

drop_column =null_var[null_var>20].keys()
drop_column
```

```python
dropped_column = data.drop(columns=drop_column)
dropped_column
```

```python
#To visualize the new dataset

sns.heatmap(dropped_column.isnull())
```

```python
#drops the rows having missing values

drop_remain_miss_val = dropped_column.dropna()
drop_remain_miss_val.shape
```

```python
sns.heatmap(drop_remain_miss_val.isnull())
```

```python
#To see if there is any null value in numeric form

drop_remain_miss_val.isnull().sum()
```

```python
#To check how many null value is present

drop_remain_miss_val.isnull().sum().sum()
```
